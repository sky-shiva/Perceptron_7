```python
import cv2
import os

output_folder = "extracted_frames"
os.makedirs(output_folder, exist_ok=True)


video_path = r'C:\Perceptron7\WIN_20260226_19_39_21_Pro.mp4'
vidcap = cv2.VideoCapture(video_path)


if not vidcap.isOpened():
    print("Error: Could not open video")
    exit()


fps = vidcap.get(cv2.CAP_PROP_FPS)  
total_frames = int(vidcap.get(cv2.CAP_PROP_FRAME_COUNT))
duration = total_frames / fps

print(f"Video FPS: {fps}")
print(f"Total frames: {total_frames}")
print(f"Video duration: {duration:.2f} seconds")


frame_interval = int(fps * 2)  
print(f"Taking 1 photo every {frame_interval} frames (every 2 seconds)")

count = 0
saved_count = 0

while True:
   
    success, image = vidcap.read()
    
    
    if not success:
        break
    
    
    if count % frame_interval == 0:
        
        timestamp = count / fps
        minutes = int(timestamp // 60)
        seconds = timestamp % 60
        
        frame_filename = os.path.join(output_folder, f"frame_{saved_count:04d}_at_{minutes}m{seconds:.1f}s.jpg")
        cv2.imwrite(frame_filename, image)
        print(f"Saved: {frame_filename} (at {timestamp:.1f} seconds)")
        saved_count += 1
    
    count += 1


vidcap.release()
print(f"Done! Extracted {saved_count} frames (1 every 2 seconds) to {output_folder}")
```
